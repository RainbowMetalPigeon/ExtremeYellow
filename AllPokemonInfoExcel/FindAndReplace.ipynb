```python
inputFolder = '/Users/Michael/Desktop/Core/Giochi/Pokémon/DisassemblyYellow/ExtremeYellow/data/wild/maps/night/'
inputFolder = '/Users/Michael/Desktop/Core/Giochi/Pokémon/DisassemblyYellow/ExtremeYellow/data/wild/maps/sevii/night/'

testFile = 'CeladonCity.asm'
inputFile = inputFolder + testFile
```

```python
search_text = 'WildMons:'
replace_text = 'WildMonsNight:'
```

```python
def ReplaceStringInFile(fileToEdit, search_text=search_text, replace_text=replace_text):
    with open(fileToEdit, 'r') as file:
        data = file.read()
        data = data.replace(search_text, replace_text)

    with open(fileToEdit, 'w') as file:
        file.write(data)
        
    return
```

```python
for entry in os.scandir(inputFolder):  
    if entry.is_file():  # check if it's a file
        localPath = entry.path
        if '.asm' in localPath:
            print(localPath)
            ReplaceStringInFile(localPath)
```
